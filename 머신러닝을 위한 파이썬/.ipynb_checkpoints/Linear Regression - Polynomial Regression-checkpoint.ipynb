```python
def f(size):
    x=np.linspace(0,5,size)
    y=x*np.sin(x**2)+1
    return (x,y)

def sample(size):
    x=np.linspace(0,5,size)
    y=x*np.sin(x**2) + 1 + np.random.randn(x.size)*0.5
    return (x,y)
```

```python
f_x, f_y = f(1000)
plt.plot(f_x, f_y)
X, y = sample(1000)
plt.scatter(X, y, s=3, c="black")
plt.show()
```

```python
X=X.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(X,y)
```

```python
f_x, f_y = f(1000)
plt.plot(f_x, f_y)
plt.scatter(X.flatten(), y.flatten(), s=3, c="black")
plt.plot(X.flatten(), lr.predict(X).flatten())
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features=PolynomialFeatures(2)
x_poly=poly_features.fit_transform(X)
x_poly[:5]
```

```python
lr=LinearRegression(fit_intercept=False)
lr.fit(x_poly,y)
```

```python
f_x, f_y = f(1000)
plt.plot(f_x, f_y)
plt.scatter(X.flatten(), y.flatten(), s=3, c="black")
plt.plot(X.flatten(), lr.predict(x_poly).flatten())
plt.show()
```

```python
def rmse(predicitons, target):
    return np.sqrt((predicitons-target)**2).mean()
```

```python
poly_range=list(range(10,50))
rmse_lr_list=list()
rmse_lasso_list=list()
rmse_ridge_list=list()

from sklearn.linear_model import Ridge, Lasso
```

```python
for poly_value in poly_range:
    poly_features=PolynomialFeatures(degree=poly_value)
    
    x_poly=poly_features.fit_transform(X)
    
    lr=LinearRegression()
    lr.fit(x_poly,y)
    
    rmse_lr_list.append(rmse(lr.predict(x_poly),y))
    
    lasso=Lasso()
    lasso.fit(x_poly,y)
    rmse_lasso_list.append(rmse(lasso.predict(x_poly),y))
    
    ridge=Ridge()
    ridge.fit(x_poly,y)
    rmse_ridge_list.append(rmse(ridge.predict(x_poly),y))
```

```python
import pandas as pd

data={'poly_range':poly_range,
     'lr_rmse':rmse_lr_list,
     'lasso_rmse':rmse_lasso_list,
     'ridge_rmse': rmse_ridge_list}

df=pd.DataFrame(data).set_index('poly_range')
df
```

```python
df.min()
```

```python
df['ridge_rmse'].sort_values().head()
```

```python
poly_features=PolynomialFeatures(23)
x_poly=poly_features.fit_transform(X)

ridge=Ridge()
ridge.fit(x_poly,y)
```

```python
f_x, f_y = f(1000)
plt.plot(f_x, f_y)
plt.scatter(X.flatten(), y.flatten(), s=3, c="black")
plt.plot(X.flatten(), ridge.predict(x_poly).flatten())
plt.show()
```

```python
plt.plot(poly_range, df['ridge_rmse'], label='ridge')
plt.plot(poly_range, df['lasso_rmse'], label='lasso')
plt.plot(poly_range, df['lr_rmse'], label='lr')
plt.legend(loc='lower right')
plt.show()
```
