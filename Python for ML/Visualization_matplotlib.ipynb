## 파이썬의 대표적인 시각화 도구인 matplotlib 을 활용하여 데이터를 시각화하는 방법

***
**핵심 키워드**

- matplotlib
- pyplot
- line
- scatter
- bar
- histogram
- boxplot
- scatter matrix

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=range(100)
y=[value**2 for value in x]
plt.plot(x,y)
plt.show()
```

```python
x=range(100)
y=range(100)
plt.plot(x,y)
plt.show()
```

- Graph 는 원래 figure 객체에 생성됨
- pyplot 객체 사용시, 기본 figure에 그래프가 그려짐

```python
x_1=range(100)
y_1=[np.cos(value) for value in x]

x_2=range(100)
y_2=[np.sin(value) for value in x]

plt.plot(x_1,y_1)
plt.plot(x_2,y_2)
plt.show() # 메모리 flush
```

```python
# 여러개의 그림 보여주기

fig=plt.figure()
fig.set_size_inches(10,5)
ax_1=fig.add_subplot(1,2,1)
ax_2=fig.add_subplot(1,2,2)

ax_1.plot(x_1,y_1,c='b')
ax_2.plot(x_2,y_2,c='g')
plt.show()
```

Set color

- color 속성을 사용
- Float -> 흑백, rgb color, predefined color 사용

```python
x_1=range(100)
x_2=range(100)

y_1=range(100)
y_2=[(value+100) for value in x_2]

plt.plot(x_1,y_1,color="#eeefff")
plt.plot(x_2,y_2, color='r')
plt.show()
```

Set linstyle

- ls 또는 linestyle 속성 사용

```python
plt.plot(x_1,y_1,c='b', linestyle='dashed')
plt.plot(x_2,y_2,c='r', ls='dotted')
plt.show()
```

Set title

- pyplot 에 title 함수 사용, figure의 subplot 별 입력 가능
- Latex 타입의 표현도 가능 (수식 표현 가능)

```python
plt.plot(x_1,y_1,c='b', linestyle='dashed')
plt.plot(x_2,y_2,c='r', ls='dotted')

plt.title('Two lines')
plt.show()
```

```python
plt.plot(x_1,y_1,c='b', linestyle='dashed')
plt.plot(x_2,y_2,c='r', ls='dotted')

plt.title('$y = \\frac{ax+b}{test}$')
plt.show()
```

Set legend

- Legend 함수로 범례를 표시함, loc 위치 등 속성 지정

```python
plt.plot(x_1,y_1,c='b', linestyle='dashed', label='line_1')
plt.plot(x_2,y_2,c='r', ls='dotted', label='line_2')

plt.legend(shadow=True, fancybox=True, loc='lower right')
```

Set grid & xylim

- Graph 보조선을 긋는 grid 와 xy 축 범위 한계를 지정

```python
plt.grid(True, lw=0.4, ls='--',c='.90')
plt.xlim(-100,200)
plt.ylim(-200,200)
```

Scatter

- scatter 함수 사용, marker : scatter 모양지정

```python
data_1= np.random.rand(512,2)
data_2=np.random.rand(512,2)

plt.scatter(data_1[:,0], data_1[:,1], c='b', marker='x')
plt.scatter(data_2[:,0], data_2[:,1],c='r', marker="^")
plt.show()
```

Scatter

- s : 데이터의 크기 지정, 데이터의 크기 비교 가능

```python
N=50

x=np.random.rand(N)
y=np.random.rand(N)

colors=np.random.rand(N)

area=np.pi*(15*np.random.rand(N))**2
plt.scatter(x,y,s=area, c=colors, alpha=0.5)
plt.show()
```

Bar Chart

- Bar 함수 사용

```python
data=[[5.,25.,50.,20.],
     [4.,23.,51.,17],
     [6.,22.,52.,19]]

X=np.arange(4)

plt.bar(X+0.00, data[0], color='b', width=0.25)
plt.bar(X+0.25, data[1], color='g', width=0.25)
plt.bar(X+0.50, data[2], color='r', width=0.25)
plt.xticks(X+0.25, ("A","B","C","D"))
plt.show()
```

```python
data=[[5.,25.,50.,20.],
     [4.,23.,51.,17],
     [6.,22.,52.,19]]

data=np.array(data)

olor_list=['b','g','r']
data_label=['A','B','C']

X=np.arange(data.shape[1])

for i in range(data.shape[0]):
    plt.bar(X, data[i], bottom=np.sum(data[:i], axis=0),
           color=color_list[i], label=data_label[i])
plt.legend()
plt.show()
```
